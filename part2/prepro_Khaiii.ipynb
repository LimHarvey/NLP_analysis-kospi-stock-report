```python
f = open('/content/drive/MyDrive/Pjt2_증권사레포트/Code/600179.txt', 'r', encoding='utf8')
text = f.readlines()
```

```python
from khaiii import KhaiiiApi

khaiii = KhaiiiApi()

class Khaiii_App :
    def __init__(self) :
        self=self
    
    def khaiii_analyze(self, words) :

        textset_mer = []

        for word in khaiii.analyze(words) :
            textset = []
            temp = str(word).split('\t')
            textset.append(temp[0])
            for i in range(len(temp[1].split('+'))) :
                textset.append(temp[1].split('+')[i]) 

            textset_mer.append(textset)
        return textset_mer

    def khaiii_corpus(self, words) :
        divided_list = self.khaiii_analyze(words)
        noun_list = []
        
        for i in divided_list :
            flag = True
            
            for j in i :
#                 print(j)
                if flag :
                    flag=False
                    continue
#                 print(j.split('/')[0] , ',' ,j.split('/')[-1])
                
                if len(j.split('/')[-1])==0 :
                    continue
                
                if j.split('/')[-1][0]==' ' :
                    continue
                    
                if j.split('/')[-1][0]=='N':
#                     print('print',j.split('/')[0])
                     noun_list.append(j.split('/')[0])
#                     noun_list.append(j)
        return noun_list
```

```python
khaiii_app = Khaiii_App()
word = khaiii_app.khaiii_corpus(text[0])

print(word)
```

```python
!pip install wordcloud
!pip install konlpy

!apt-get update -qq
!apt-get -qq -y install fonts-nanum* 

from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt

counts = Counter(word)
tags = counts

wc = WordCloud(font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf', width=600, height=400, background_color='white')
wc.generate_from_frequencies(dict(tags))

plt.figure(figsize=(20,13))
plt.axis('off')
plt.imshow(wc)
plt.show()

wc.to_file('khaiii_analyze.png')
```
