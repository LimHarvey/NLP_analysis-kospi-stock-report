```python
import re
```

```python
# 제목에서 숫자 6자리(code)만 추출
for n, i in enumerate(df.제목):
    s = i
    m = re.search('[0-9]{6}',s)
    print(n,m.group())
    
# 70 번째 데이터에서 에러 발견
```

```python
# 전체 데이터 error개수와 error발생 인덱스 추출

count = 0
error_index = []
for n, i in enumerate(df.제목):
    try : 
        s = i
        # 제목에서 숫자 6자리(code)만 추출
        m = re.search('[0-9]{6}',s)
        print(n,m.group())
        
    except Exception as e:
        print(n, i,'error:',e)
        error_index.append(n)
        count += 1
        pass

print()
print('총 error 개수 :', count)
```
