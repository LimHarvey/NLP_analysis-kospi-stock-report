### PDF Parsing & preprocessing

```python
# 증권사별 전처리 함수
# Compliance Notice 
import re
def Securities_Co(pdf_name, txt_content):
    idx = pdf_name.split('_')[-1].replace('.pdf','')
    search_df = df[df['uri'].str.contains(idx)]
    source_co = search_df['제공출처'].values[0]
    global pre_txt
    if source_co == '유안타증권' :
        txt_content = re.split('이 자료에 게재된', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    elif source_co == '현대증권' :
        txt_content = re.split('투자자 고지사항 | 투자자 고지 사항', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    elif source_co == 'NH투자증권' :
        txt_content = re.split('종목 투자등급', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    elif source_co == '미래에셋대우' :
        txt_content = re.split('투자의견 분류', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    elif source_co == '미래에셋증권' :
        txt_content = re.split('Recommendations', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    elif source_co == 'KB증권' :
        txt_content = re.split('고지 사항', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    elif source_co == '하이증권' :
        txt_content = re.split('당 보고서 공표일', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    elif source_co == '하이투자증권' :
        txt_content = re.split('당 보고서 공표일', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    elif source_co == '하나금융투자' :
        txt_content = re.split('투자등급 관련사항', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    elif source_co == '신한금융' :
        txt_content = re.split('이 자료에 게재된', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
    else :
        txt_content = re.split(' Compliance | compliance ', txt_content, maxsplit=1)
        pre_txt = txt_content[0]
 
```

```python
import time
import os
import re
from tika import parser

def PDF_Parsing(pdf_list_path):
    start = time.time()
    if not os.path.isdir('./text2/'):
        os.mkdir('./text2/')

    # pdf파일이 있는 경로를 반복list로 설정
    pdf_list = os.listdir(pdf_list_path)
    for file in pdf_list: # pdf파일 경로에 있는 파일
        text = parser.from_file(pdf_list_path + '{}'.format(file)) 
        text['content']=str(text['content']).replace('\n', '')
        content = text['content']
        # 증권사별 전처리 함수
        Securities_Co(file, content)
        file = file.replace('.pdf','')
        with open('./text2/{}.txt'.format(file), 'w', encoding='utf-8') as f:
            f.write(pre_txt)

    print("time :", time.time() - start)
```

```python
# pdf 파일 경로 지정
# 200개 41초
# 2000개 465초
PDF_Parsing('./pdf1/')
```
